### Load Libraries

```python
from A_Data_loader import dataframes
import pandas as pd
```

### Extract Tables

```python
customers = dataframes["olist_customers_dataset"]

orders = dataframes["olist_orders_dataset"]

payments = dataframes["olist_order_payments_dataset"]

products = dataframes["olist_products_dataset"]

reviews = dataframes["olist_order_reviews_dataset"]

items = dataframes["olist_order_items_dataset"]
```

### Fix Date Columns

```python
date_columns = [
    "order_purchase_timestamp",
    "order_approved_at",
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date"
]

for col in date_columns:

    orders[col] = pd.to_datetime(
        orders[col],
        errors="coerce"
    )

    print(orders.dtypes)
```

### Fix Numeric Columns

```python
numeric_cols = [
    "product_weight_g",
    "product_length_cm",
    "product_height_cm",
    "product_width_cm"
]

for col in numeric_cols:

    products[col] = pd.to_numeric(
        products[col],
        errors="coerce"
    )

# Payments Table 

payments["payment_value"] = pd.to_numeric(
    payments["payment_value"],
    errors="coerce"
)
```

### Missing Value Treatment

```python
print(products.isnull().sum())
```

```python
products["product_category_name"] = (
    products["product_category_name"]
    .fillna("Unknown")
)
```

```python
for col in numeric_cols:

    products[col] = products[col].fillna(
        products[col].median()
    )
```

```python
reviews["review_comment_title"] = (
    reviews["review_comment_title"]
    .fillna("No Title")
)

reviews["review_comment_message"] = (
    reviews["review_comment_message"]
    .fillna("No Review")
)
```

### Remove Duplicate Rows

```python
for name, df in dataframes.items():

    before = len(df)

    df.drop_duplicates(inplace=True)

    after = len(df)

    print(
        f"{name}: Removed {before-after} duplicates"
    )
```
